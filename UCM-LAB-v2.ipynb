```python
#PHSX 216 Projectile Motion Lab plotting code asdfadsfadsf
#performs 2nd degree polynomial fit
#Nathan Johnson
#Updated October 8 2024

import numpy as np
import matplotlib.pyplot as plt

#-----------------------------------------------------------------------#


#DATA TO BE PLOTTED

Tsquared = np.array([821.48,483.62,345.91,159.69,144.47]) #N squared   <-----
t = np.array([11.076,9.546,8.3,6.338,5.764]) #seconds   <-----
err_SD = np.array([0.775,0.609,0.413,0.169,0.304])                   #                 <-----

x = t             #this should be the array you want to plot on the x axis #        <-----
y = Tsquared             #this should be the array you want to plot on the y axis #        <-----
dy = err_SD        #this should be your error in y array                    #        <-----

plot_title= 'Uniform Circular Motion'                                                                #                 <-----
x_label= 't(s)'                                                    #                 <-----
y_label= 'T^2'                                                    #                 <-----

#-----------------------------------------------------------------------#
#-----DON'T NEED TO CHANGE ANYTHING BEYOND THIS POINT!!!-----


#size the plot
plt.figure(figsize=(15,10))

#create scatter plot
plt.scatter(x, y, color='blue', marker='o')

#create labels
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(plot_title)

#fitting to a 2nd degree polynomial
c,b,a=np.polynomial.polynomial.polyfit(x,y,2,w=dy)

#Annotate with values of A, B, C from best fit polynomial
plt.annotate('best fit line: $f(x)=Ax^2+Bx+C$',(0.05, 0.95), xycoords='axes fraction')

plt.annotate('     A = {:.6f}'.format(a),(0.05, 0.9), xycoords='axes fraction')

plt.annotate('     B = {:.6f}'.format(b),(0.05, 0.85), xycoords='axes fraction')

plt.annotate('     C = {:.6f}'.format(c),(0.05, 0.8), xycoords='axes fraction')

#Create fit line
xnew = np.linspace(x.min(), x.max(), 300)
fit = a*xnew**2 + b*xnew +c

plt.plot(xnew, fit, color='green')
plt.show()
```
