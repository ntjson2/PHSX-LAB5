```python
#PHSX 216 Hooks Law Lab - Brass
#Nathan Johnson
#Updated October 2024

import numpy as np
import matplotlib.pyplot as plt

#-----------------------------------------------------------------------#
#----------UPDATE THIS SECTION!!!----------

#DATA TO BE PLOTTED

xDelta = np.array([1.5,2.3,2.2,2.5,2.5,2.5,2.5,2.8]) #what are units?   <-----
Fcap = np.array([2.68,2.87,3.07,3.26,3.46,3.66,3.85,4.05]) #what are units?   <-----
err_ycap = np.array([2.00,8.00,6.00,6.00,6.00,7.00,7.00,8.00])                   #                 <-----

x = xDelta             #this should be the array you want to plot on the x axis #        <-----
y = Fcap             #this should be the array you want to plot on the y axis #        <-----
dy = err_ycap        #this should be your error in y array                    #        <-----

plot_title= 'Steel spring calculations'                                                                #                 <-----
x_label= 'Change in X'                                                    #                 <-----
y_label= 'Force (F)'                                                    #                 <-----

#-----------------------------------------------------------------------#
#-----DON'T NEED TO CHANGE ANYTHING BEYOND THIS POINT!!!-----


#size the plot
plt.figure(figsize=(15,10))

#create scatter plot
plt.scatter(x, y, color='blue', marker='o')

#create labels
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(plot_title)

#fitting to a 2nd degree polynomial
c,b,a=np.polynomial.polynomial.polyfit(x,y,2,w=dy)

#Annotate with values of A, B, C from best fit polynomial
plt.annotate('best fit line: $f(x)=Ax^2+Bx+C$',(0.05, 0.95), xycoords='axes fraction')

plt.annotate('     A = {:.6f}'.format(a),(0.05, 0.9), xycoords='axes fraction')

plt.annotate('     B = {:.6f}'.format(b),(0.05, 0.85), xycoords='axes fraction')

plt.annotate('     C = {:.6f}'.format(c),(0.05, 0.8), xycoords='axes fraction')

#Create fit line
xnew = np.linspace(x.min(), x.max(), 300)
fit = a*xnew**2 + b*xnew +c

plt.plot(xnew, fit, color='green')
plt.show()
```
